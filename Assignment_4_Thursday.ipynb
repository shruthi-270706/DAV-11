```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Year': [1949]*12 + [1950]*12,
    'Month': list(range(1,13)) * 2,
    'Passengers': [112,118,132,129,121,135,148,148,136,119,104,118,
                   115,126,141,135,125,149,170,170,158,133,114,140]
}
df = pd.DataFrame(data)
pivot_df = df.pivot_table(index='Month', columns='Year', values='Passengers')

sns.heatmap(pivot_df, annot=True, cmap="coolwarm")
plt.title("Monthly Passengers (1949-1950)")
plt.show()
```
