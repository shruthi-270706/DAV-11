```python
import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022]
revenue = [5.2, 6.0, 5.8, 6.5, 7.2]

plt.bar(years, revenue, color='skyblue')

plt.title('Annual Revenue (2018-2022)')
plt.xlabel('Year')
plt.ylabel('Revenue (in millions)')

plt.show()
```
